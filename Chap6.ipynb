```python
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Get data (x-axis and y-axis)
# Simulate ocean depth (meters) and corresponding seawater temperature (Celsius)
depth = np.linspace(0, 1000, 50)  # Depth from 0 to 1000 meters
temperature = 20 * np.exp(-depth / 300) + 2  # Simulate temperature decreasing with depth

# Step 2: Create a figure object (window)
plt.figure(figsize=(8, 6), dpi=100)  # Set figure size and resolution

# Step 3: Plot the graph using plot() function
plt.plot(depth, temperature, color='blue', linestyle='-', linewidth=2, label='Sea Temperature')

# Step 4: Set graph attributes (title, legend, text, axes, etc.)
plt.title('Ocean Temperature Profile', fontsize=14)  # Set title
plt.xlabel('Depth (m)', fontsize=12)                 # Set X-axis label
plt.ylabel('Temperature (°C)', fontsize=12)          # Set Y-axis label
plt.legend(loc='upper right')                        # Show legend
plt.grid(True, linestyle='--', alpha=0.7)            # Add grid lines
plt.text(500, 12, 'Thermocline Region', fontsize=10, color='red')  # Add text annotation

# Step 5: Load mouse/keyboard events for interactive features
# Example: Bind a key press event. Press 's' to save the figure.
def on_key(event):
    if event.key == 's':
        plt.savefig('ocean_temp_profile.png', dpi=150)
        print("Figure saved as ocean_temp_profile.png")

plt.gcf().canvas.mpl_connect('key_press_event', on_key)

# Step 6: Display or save the figure
plt.show()
```

```python
# 雷达图
import numpy as np
import matplotlib.pyplot as plt

# --- 解决中文乱码的核心配置 ---
plt.rcParams['font.sans-serif'] = [
    'Microsoft YaHei',  # Windows系统常用：微软雅黑
    'SimHei',           # Windows系统常用：黑体
    'PingFang SC',      # macOS系统常用：苹方
    'Arial Unicode MS', # macOS/通用备选
    'sans-serif'        # 保底方案
]
plt.rcParams['axes.unicode_minus'] = False  # 解决负号'-'显示为方块的问题
# ---------------------------

# 1. 准备数据
labels = np.array(['准确率(R²)', '误差控制(RMSE)', '计算速度'])
model_A_scores = np.array([9, 7, 3])
model_B_scores = np.array([3, 6, 9])
model_C_scores = np.array([8, 8, 7])

# 2. 闭合图形
angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
scores_A = np.concatenate((model_A_scores, [model_A_scores[0]]))
scores_B = np.concatenate((model_B_scores, [model_B_scores[0]]))
scores_C = np.concatenate((model_C_scores, [model_C_scores[0]]))
angles += angles[:1]

# 3. 创建极坐标画布并绘图
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

ax.plot(angles, scores_A, 'o-', linewidth=2, label='Model A (高精度)')
ax.fill(angles, scores_A, alpha=0.25)
ax.plot(angles, scores_B, 's-', linewidth=2, label='Model B (高速度)')
ax.fill(angles, scores_B, alpha=0.25)
ax.plot(angles, scores_C, 'd-', linewidth=2, label='Model C (均衡型)')
ax.fill(angles, scores_C, alpha=0.25)

# 4. 设置标签和美化
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels, fontsize=12)
ax.set_ylim(0, 10)
ax.set_title('Marine Model Performance Comparison', fontsize=14, pad=20)
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import shapely.geometry

# 1. 创建画布与地图投影
fig = plt.figure(figsize=(10, 6))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()  # 添加海岸线作为背景参考
ax.set_extent([-130, -110, 40, 60])  # 设置显示范围

# 2. 绘制一个蓝色圆点，位置 (经度, 纬度)
# 注意：Geodetic() 和 PlateCarree() 在绘制点时效果相同，但 PlateCarree 更常用
ax.plot(-120.5, 47.3, 'bo', markersize=10, transform=ccrs.PlateCarree())

# 3. 绘制一条折线
lons = np.linspace(-125, -115, 10)
lats = np.linspace(45, 55, 10)
ax.plot(lons, lats, transform=ccrs.PlateCarree(), color='blue', linewidth=2)

# 4. 绘制面（多边形）
polygon = [
    (-123, 46), 
    (-122, 54), 
    (-112, 50), 
    (-124, 46), 
    (-123, 46)  # 闭合点
]
ax.add_geometries(
    [shapely.geometry.Polygon(polygon)], 
    crs=ccrs.PlateCarree(), 
    facecolor='green', 
    alpha=0.7, 
    edgecolor='black', 
    linewidth=2
)

plt.show()
```

```python
# -*- coding: utf-8 -*-
"""
Created on  26.05.20
@author: eva
"""
import numpy as np
import xarray as xr

import matplotlib
import matplotlib.pyplot as plt
from matplotlib import rcParams

import cartopy.crs as ccrs
import cartopy.feature as cfeature
from cartopy.mpl.ticker import LongitudeFormatter, LatitudeFormatter


# 1. 读取nc文件
fidname = "OISST20160401_20260430.nc"
ds = xr.open_dataset(fidname)

# 2. 快速查看数据全貌（会显示所有变量、维度和属性）
print(ds)

# 3. 提取数据（像操作字典一样简单）
sst_data = ds['SST']
lons = ds['LON401_540']
lats = ds['LAT381_560']
times = ds['TIME'].values

# 4. 可视化数据 - 字体与格式配置
font = {
    'family': 'Times New Roman',
    'weight': 'semibold',
    'size': 14
}
matplotlib.rc("font", **font)

config = {
    "mathtext.fontset": 'stix',
}
rcParams.update(config)

# 5. 创建画布与地图投影
fig = plt.figure(figsize=(8, 7))
ax = plt.subplot(111, projection=ccrs.PlateCarree())

# 6. 设定绘图范围与刻度
extent = [110, 130, 20, 46]  # 限定绘图范围
ax.set_extent(extent)
ax.set_xticks(np.linspace(110, 130, 5), crs=ccrs.PlateCarree())
ax.set_yticks(np.linspace(20, 45, 6), crs=ccrs.PlateCarree())

lon_formatter = LongitudeFormatter(zero_direction_label=True)
lat_formatter = LatitudeFormatter()
ax.xaxis.set_major_formatter(lon_formatter)
ax.yaxis.set_major_formatter(lat_formatter)  # 使ax具有经纬度格式

# 7. 绘制SST填色图
clevs = np.linspace(0, 35, 30)
map_car = ax.contourf(
    lons, lats, sst_data[0, 0, :, :], clevs,
    transform=ccrs.PlateCarree(),
    cmap=plt.cm.jet
)  # contouf可填充轮廓线，contour单纯画轮廓线。

# 8. 添加颜色条
ticks = np.linspace(0, 35, 11)
cax = fig.add_axes([0.85, 0.095, 0.05, 0.78])  # 色标参数
cb = fig.colorbar(
    map_car, extend='both', shrink=1, orientation='vertical',
    ax=ax, ticks=ticks, cax=cax
)  # horizontal
# 设置颜色条的标签为“温度 (°C)”
cb.set_label(' °C', fontsize=12)
cb.ax.tick_params(labelsize=12)

# 9. 设置标题与地理要素
ax.set_title(f"{str(times[0])[0:10]} SST")

# 添加清晰的地理要素（陆地、海洋、高分辨率海岸线）
ax.add_feature(cfeature.LAND, facecolor='lightgray')
ax.add_feature(cfeature.OCEAN, facecolor='aliceblue')
ax.add_feature(cfeature.COASTLINE, linewidth=1.2, edgecolor='black')  # 加粗黑色海岸线
ax.gridlines(draw_labels=False, linestyle='--', color='gray', alpha=0.6)

plt.show()
```
